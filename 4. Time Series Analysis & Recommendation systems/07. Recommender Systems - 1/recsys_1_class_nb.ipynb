```python
# We are only interested in positive quantities
df = df[df['Quantity']>=0]
df.info()
```

```python
df.dropna(axis=0, subset=['InvoiceNo'],inplace=True)
df['InvoiceNo']=df['InvoiceNo'].astype('str')
df = df[~df['InvoiceNo'].str.contains('C')]
df.shape
```

```python
df = df[df['Country'] =="United Kingdom"]
```

```python
df.head()
```

```python
df.groupby(['InvoiceNo','Description'])['Quantity'].sum()
```

```python
df.groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack()
```

```python
data = df.groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().reset_index().fillna(0)
```

```python
data.head()
```

```python
data = data.set_index('InvoiceNo')
```

```python
data.head()
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

data = data.map(encode_units)
data.head()
```

```python
data = data[data.sum(axis=1) >= 2]
```

```python
len(data)
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
frequent_itemsets_plus = apriori(data, min_support=0.03,
                                 use_colnames=True).sort_values('support', ascending=False).reset_index(drop=True)

frequent_itemsets_plus['length'] = frequent_itemsets_plus['itemsets'].apply(lambda x: len(x))

frequent_itemsets_plus[frequent_itemsets_plus['length']>=2]
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules=association_rules(frequent_itemsets_plus, metric='lift',
                  min_threshold=0.4).sort_values('lift', ascending=False).reset_index(drop=True)
association_rules.head(5)
```

```python
# Filtering association rules where leverage is greater than 0.03
leverage_filtered = association_rules[association_rules['leverage'] > 0.03]
display(leverage_filtered.head())
```
