```python
from google.colab import userdata

aws_access_key_id = userdata.get('aws_access_key_id')
aws_secret_access_key = userdata.get('aws_secret_access_key')
EndpointName = userdata.get('EndpointName')
```

```python
import boto3
import json
```

```python
# km_driven	mileage	age	Petrol	Diesel	Electric
tst="50000,20,5,1,0,0"
```

```python
runtime = boto3.Session().client('sagemaker-runtime',region_name='us-west-2',
                  aws_access_key_id=aws_access_key_id,
                  aws_secret_access_key=aws_secret_access_key )
```

```python
response = runtime.invoke_endpoint(
    EndpointName=EndpointName, ContentType="text/csv", Body=tst
)
```

```python
result = response["Body"].read().decode("ascii")
print(result)
```

```python
payload = "5000,20,5,1,0,0"
response = runtime.invoke_endpoint(
    EndpointName=EndpointName,
    ContentType="text/csv",
    Body=payload
)
result = response["Body"].read().decode("utf-8")
print(result)
```
